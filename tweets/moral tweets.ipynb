```python
import glob
import pandas as pd
import numpy as np
import re
import os

path = "/home/linda/manifestos/tweets/downloads/coded/"
allfiles = glob.glob(os.path.join(path,"*.csv"))

peeps = [p.split('/')[-1].strip('.csv.') for p in alltweets]

names = [pp.split('_')[-3].strip() for pp in peeps]
parties = [pp.split('_')[-2].strip() for pp in peeps]

lidmaatschap = dict(zip(names,parties))
namefile = dict(zip(names,allfiles))

```

```python
np_array_list = []
data = []
for name in names:
    file_= namefile[name]
    df = pd.read_csv(file_,index_col=None, header=0)
    df['politicians'] = name
    df['party'] = lidmaatschap[name]
    data.append(df)
    np_array_list.append(df.as_matrix())

comb_np_array = np.vstack(np_array_list)
big_frame = pd.DataFrame(comb_np_array)

big_frame.columns = ['id','tweet','date', 'likes', 'retweets', 'morallist', 'harmvirtuelist', 'harmvicelist', 'harmlist', 'fairnessvirtuelist', 'fairnessvicelist',  'fairnesslist', 'ingroupvirtuelist', 'ingroupvicelist', 'ingrouplist', 'purityvirtuelist', 'purityvicelist', 'puritylist', 'authorityvirtuelist', 'authorityvicelist', 'authoritylist', 'length', 'name', 'party']

bf = big_frame.fillna(value = 0)


```

```python
parties = pd.get_dummies(bf['party'])

bfd = pd.concat([bf, parties], axis=1)  

#bf['com'] = bf.sf
#bf['cons'] = bf.uup + bf.con
bfd['con'] = bfd.con
bfd['eco'] = bfd.gpew
#bfd['eth'] = bfd.snp + bfd.pc + bfd.dup
bfd['eth'] = bfd.snp + bfd.dup
bfd['lib'] = bfd.libdem
bfd['sip'] = bfd.ukip
#bfd['soc'] = bfd.labour + bfd.sdlp
bfd['soc'] = bfd.labour


bfj = bfd.drop_duplicates(subset='id')

# data is nu goed geladen
# nu nog analyseren
# eigenlijk wil ik kijken met describe of bepaalde foundations meer voorkomen bij bepaalde partijen
# is dus met groupby
# ook logreg ipv ols
# sentiment ook van belang om mee te nemen
# mlm lijkt logischer ook als er meer mps bijkomen

# wellicht twitteraccounts meenemen van partijen en van eu-politici (dan heb ik ukip ook meteen)
```

```python
from statsmodels.formula.api import ols

myfittedregression = ols(formula = 'morallist ~ length + eco + eth + lib + sip + soc ', data = bfj).fit()
print(myfittedregression.summary())

myfittedregression = ols(formula = 'harmlist ~ length + eco + eth + con + sip + soc ', data = bfj).fit()
print(myfittedregression.summary())

myfittedregression = ols(formula = 'authoritylist ~ length + eco + eth + con + sip + soc ', data = bfj).fit()
print(myfittedregression.summary())

myfittedregression = ols(formula = 'ingrouplist ~ length + eco + eth + con + sip + soc ', data = bfj).fit()
print(myfittedregression.summary())

myfittedregression = ols(formula = 'puritylist ~ length + eco + eth + con + sip + soc ', data = bfj).fit()
print(myfittedregression.summary())


```
